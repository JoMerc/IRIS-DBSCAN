```python
import pandas as pd
from sklearn import datasets
```

```python
data = datasets.load_iris()
```

```python
X = pd.DataFrame(data.data,
                 columns=list(data.feature_names))
```

```python
X.head(5)
```

```python
from sklearn import cluster
```

```python
model = cluster.DBSCAN(eps=0.8)
```

```python
results = model.fit(X)
```

```python
X['cluster'] = results.labels_
X['target'] = data.target
X['cl'] = 'lookatme'
X.head(10)
```

```python
classification_result = X[['cluster', 'target', 'cl']]\
                            .groupby(['cluster', 'target'])\
                            .agg('count')
```

```python
classification_result
```
